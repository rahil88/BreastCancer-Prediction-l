```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
import numpy as np
```

```python
# Importing dataset
df = pd.read_csv('BreastCancer.csv')
df
```

```python
benign_data = df[df['Class'] == 'benign']
```

```python
benign_data.isnull().sum()
```

```python
# Replace missing values in the 'Bare.nuclei' column with the mean
benign_data['Bare.nuclei'].fillna(benign_data['Bare.nuclei'].mean(),inplace=True)

benign_data
```

```python
malignant_data = df[df['Class'] == 'malignant']
malignant_data
```

```python
# Replace missing values in the 'Bare.nuclei' column with the mean
malignant_data['Bare.nuclei'].fillna(malignant_data['Bare.nuclei'].mean(),inplace=True)

malignant_data
```

```python
df = pd.concat([benign_data,malignant_data], axis=0)
```

```python
# Check for null values in all columns
df.isnull().sum()
```

```python
# There is no need of encoding diagnosis column as it is the dependent variable(y)
# But we encode the diagnosis column to find correlation and make a heatmap
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Class_enc'] = le.fit_transform(df['Class'])
```

```python
# Deleting the diagnosis column as we encoded it in diagnosis_enc
df = df.drop('Class', axis=1)
df
```

```python
# Plotting a heatmap
x = df.corr()
plt.figure(figsize=(30, 20))
sns.heatmap(x,annot=True)
```

```python
# Finding the correlation of all the columns with the diagnosis column
correlations = df.corr()['Class_enc'].abs().sort_values()

# Print the correlations
print(correlations)
```

```python
df1 = df.drop('Id',axis=1)
```

```python
# Apply PCA to reduce dimensionality
# pca = PCA(n_components=0.95)  # Retain 95% of variance
# X_pca = pca.fit_transform(X)

# Get the list of removed features
# removed_features = []
# for i in range(len(X.columns)):
#     if i >= X_pca.shape[1]:
#         removed_features.append(X.columns[i])
# print("Removed features:", removed_features)
```

```python
# Defining the independent variable (X) and dependent variable (y)
X = df1.drop(["Class_enc"], axis = 1)
y = df1["Class_enc"]

# Splitting the dataset into 70% training and 30% testing data with use of stratified sampling method
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X, y, test_size=0.3, stratify=y)
#stratify ensures that the entire data is balaced and not just based on one class eg.just benign data entirely or vice versa
```

```python
#importing necessary libraries for model
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
```

```python
X=df.iloc[:,0:9] #dividing independent features i.e. X
y=df.iloc[:,9] #dividing dependent feature i.e. target variable y
```

```python
#scaling the data to bring values in range
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
#training an Artificial Neural Network (ANN) model
model=MLPClassifier(hidden_layer_sizes=(100, 50),max_iter=500,activation='relu',solver='adam',random_state=42)
model.fit(X_train_scaled,y_train)
```

```python
#Testing the model and determine accuracy
y_pred=model.predict(X_test_scaled)
accuracy=accuracy_score(y_test, y_pred)
print("Accuracy of the classifier:", accuracy)
```

```python
# Making a Confusion Matrix to check the FP and FN 
from sklearn.metrics import confusion_matrix
cm_lr = confusion_matrix(y_test, y_pred)
cm_lr
```

```python
# Bagging using RandomForest Classifier
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(n_estimators=100, random_state=42)
random_forest.fit(X_train, y_train)
y_pred_rf = random_forest.predict(X_test)
print(f"Random Forest Accuracy: {accuracy_score(y_test, y_pred_rf):.3f}")
```
